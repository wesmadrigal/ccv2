```python
from bs4 import BeautifulSoup
import urllib2
import re
```

```python
soup = BeautifulSoup( urllib2.urlopen('http://www.craigslist.org').read() )
```

```python
soup
```

```python
# get all of the children and build a mapping
children = [ i.findChild('a') for i in soup.findAll('li') ]
seq = re.compile(r'[a-z]+')
mappings = { child.string : child.get('href') for child in children if child is not None }
```

```python
import json
```

```python
f = open('../mappings.txt', 'w')
f.write(json.dumps(mappings))
f.close()
```
